```python
def missingCounter(df):
    missing_data = df.isna().sum()
    missing_data = missing_data[missing_data > 0].sort_values(ascending = False )
    print(missing_data )
    return missing_data.index
missingCounter(hotel_df)
```

```python
# Xóa các dòng có giá trị NaN trong cột 'Name'
hotel_df = hotel_df.dropna(subset=['Name'])

# Hàm chuyển chuỗi có dấu sang không dấu
def remove_diacritics(text):
    if pd.isna(text): return ''
    text = unicodedata.normalize('NFD', str(text))
    text = ''.join(c for c in text if unicodedata.category(c) != 'Mn')
    text = text.replace('Đ', 'D').replace('đ', 'd')  # Xử lý đặc biệt
    return text


# Làm sạch cột 'Name'
def clean_name(text):
    if pd.isna(text): return ''
    text = remove_diacritics(text)
    text = text.replace('Unclaimed', '').strip()
    return text

# Làm sạch địa chỉ: bỏ dấu + bỏ mã số đứng trước tên quốc gia
def clean_address(text):
    if pd.isna(text): return ''
    text = remove_diacritics(text)

    # Bước 1: Xóa mã vùng kiểu "20000", "310000", v.v. trước tên quốc gia
    pattern = r'\s\d{4,6}(?=\s*(Vietnam|Thailand|Philippines|Malaysia|Singapore|Indonesia|Cambodia|Laos|Myanmar|Brunei|Timor))'
    text = re.sub(pattern, '', text, flags=re.IGNORECASE)

    # Bước 2: Thêm dấu "," nếu không có dấu phẩy ngay trước tên quốc gia
    text = re.sub(r'(?<![,])\s+(?=(Vietnam|Thailand|Philippines|Malaysia|Singapore|Indonesia|Cambodia|Laos|Myanmar|Brunei|Timor)$)', r', ', text.strip(), flags=re.IGNORECASE)

    return text.strip()


hotel_df['Name'] = hotel_df['Name'].apply(clean_name)
hotel_df['Address'] = hotel_df['Address'].apply(clean_address)
```

```python
hotel_df['Phone Number'] = hotel_df['Phone Number'].fillna("Unknown phone number")
hotel_df['Website'] = hotel_df['Website'].fillna("Unknown Website")
hotel_df['Description'] = hotel_df['Description'].fillna("Unknown Description")
```

```python
hotel_df = hotel_df.drop(columns=['Near Hotels', 'Hotels Group'])
```

```python
def extract_category(ranking_str):
    if isinstance(ranking_str, str):
        match = re.match(r'#\d+\s+of\s+\d+\s+(.+?)\s+in\s+.+', ranking_str)
        if match:
            return match.group(1).strip()
    return 'Unknown'

def parse_ranking(ranking):
    if isinstance(ranking, str):
        match = re.match(r'#(\d+)\s+of\s+(\d+)\s+(.+)', ranking)
        if match:
            return int(match[1]), int(match[2]), match[3]
    return None, None, None

# Áp dụng và đổi tên đúng theo yêu cầu
hotel_df[['ranking_position', 'ranking_total', 'category']] = hotel_df['Ranking'].apply(
    lambda x: pd.Series(parse_ranking(x))
)

# Xử lý dữ liệu thiếu
hotel_df['ranking_position'] = hotel_df['ranking_position'].fillna(-1).astype(int)
hotel_df['ranking_total'] = hotel_df['ranking_total'].fillna(-1).astype(int)
hotel_df['category'] = hotel_df['Ranking'].apply(extract_category)
```

```python
# Đảm bảo cột Price đã được chuẩn hóa float
hotel_df['Price'] = hotel_df['Price'].replace('[\$,]', '', regex=True).astype(str)
hotel_df['Price'] = hotel_df['Price'].replace('nan', np.nan).astype(float)
```

```python
hotel_df['Price'] = hotel_df['Price'].fillna("Unknown price")
```

```python
hotel_df['Near Restaurants'] = hotel_df['Near Restaurants'].fillna(0).astype(int)
hotel_df['Near Attractions'] = hotel_df['Near Attractions'].fillna(0).astype(int)
hotel_df['Walker Score'] = hotel_df['Walker Score'].fillna(0).astype(int)
```

```python
def clean_airport_pair(name_str, dist_str):
    if pd.isna(name_str) or pd.isna(dist_str):
        return name_str, dist_str  # giữ nguyên nếu thiếu hẳn 1 bên

    names = [n.strip() for n in str(name_str).split(',')]
    dists = [d.strip() for d in str(dist_str).split(',')]

    # Trường hợp số phần tử không khớp
    if len(names) != len(dists):
        return name_str, dist_str  # không xử lý

    cleaned_pairs = []
    for n, d in zip(names, dists):
        # loại nếu là "Rental cars" và dist là "nan"
        if n.lower() == 'rental cars' and d.lower() == 'nan':
            continue
        cleaned_pairs.append((n, d))

    if not cleaned_pairs:
        return None, None  # cả 2 đều bị loại

    new_names, new_dists = zip(*cleaned_pairs)
    return ', '.join(new_names), ', '.join(new_dists)

# Áp dụng xử lý vào DataFrame
airport_cleaned = hotel_df.apply(
    lambda row: clean_airport_pair(row['Airport Names'], row['Airport Distances']),
    axis=1
)

# Gán kết quả đã clean lại
hotel_df['Airport Names'] = airport_cleaned.apply(lambda x: x[0])
hotel_df['Airport Distances'] = airport_cleaned.apply(lambda x: x[1])



```

```python
# Gán giá trị mặc định nếu thiếu
hotel_df['Airport Names'] = hotel_df['Airport Names'].fillna('Unknown Airport')
hotel_df['Airport Distances'] = hotel_df['Airport Distances'].fillna('Unknown Distance')

def extract_miles(distance_str):
    if isinstance(distance_str, str):
        match = re.search(r'([\d.]+)\s*mi', distance_str.lower())
        if match:
            return float(match.group(1))
    return "Unknown Distance"

# Tạo cột số hóa
hotel_df['airport_distance_miles'] = hotel_df['Airport Distances'].apply(extract_miles)
```

```python
country_language_map = {
    'Vietnam': ['Vietnamese'],
    'Thailand': ['Thai'],
    'Laos': ['Lao'],
    'Cambodia': ['Khmer'],
    'Indonesia': ['Indonesian'],
    'Malaysia': ['Malay'],
    'Singapore': ['English'],
    'Philippines': ['Filipino'],
    'Myanmar': ['Burmese']
}

def extract_languages(lang, country):
    if isinstance(lang, str) and lang.lower() not in [np.nan, 'nan', 'unknown', '']:
        return [l.strip() for l in lang.split(',')]
    # Nếu lang bị thiếu thật sự → dùng country
    return country_language_map.get(country, ['Unknown'])


hotel_df['Languages Spoken'] = hotel_df.apply(
    lambda row: extract_languages(row['Languages Spoken'], row['Country']),
    axis=1
)
```

```python
def extract_stars(text):
    if isinstance(text, str):
        match = re.search(r'(\d+(\.\d+)?)', text)
        if match:
            return float(match.group(1))
    return "No stars" 

hotel_df['Hotel Class'] = hotel_df['Hotel Class'].apply(extract_stars)
```

```python
hotel_df['Room Features List'] = hotel_df['Room Features'].apply(
    lambda x: [f.strip() for f in str(x).split(',')] if pd.notna(x) else []
)
```

```python
hotel_df['Room Types List'] = hotel_df['Room Types'].apply(
    lambda x: [t.strip() for t in str(x).split(',')] if pd.notna(x) else []
)
```

```python
hotel_df['Hotel Style List'] = hotel_df['Hotel Style'].apply(
    lambda x: [s.strip() for s in str(x).split(',')] if pd.notna(x) else []
)
```

```python
hotel_df = hotel_df.drop(columns=['Room Features', 'Room Types', 'Hotel Style', 'Ranking', 'Airport Distances'])
```

```python
hotel_df['Overall Rating Value'] = hotel_df['Overall Rating Value'].fillna(-1).astype(float)
hotel_df['Overall Rating Text'] = hotel_df['Overall Rating Text'].fillna('Unknown')

review_cols = [
    'Terrible Reviews', 'Poor Reviews', 'Average Reviews',
    'Very Good Reviews', 'Excellent Reviews', 'Overall Review Count'
]

for col in review_cols:
    hotel_df[col] = (
        hotel_df[col]
        .astype(str)
        .str.replace(',', '', regex=False)
        .str.replace(' reviews', '', regex=False)
        .str.replace(' review', '', regex=False)
        .replace(['nan', 'None', ''], np.nan)
        .astype(float)   # 👉 Thêm bước này trước
        .fillna(0)
        .astype(int)     # 👉 Ép kiểu int cuối cùng
    )
```

```python
hotel_df.to_csv('Hotel_Data\Hotel_Tripadvisor_Data_SEA_cleanedforweb.csv', index=False, encoding='utf-8', quoting=1)
```
